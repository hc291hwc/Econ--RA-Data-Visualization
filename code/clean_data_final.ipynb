```python
import pandas as pd 
import numpy as np 
```

```python
df_final_03_to_10 = pd.DataFrame(columns = ['occupation', 'total', 'men', 'women', 'race', 'year'])

for year in range(2003, 2011):
    data = pd.read_csv(f'aat10_{year}.txt', sep='[ \t]{2,}', header=None, error_bad_lines=False)
    data.replace(r"^ +| +$", r"", regex=True, inplace=True)
    data[0] = data[0].str.rstrip('.')
    col_list = ['occupation',
            f'{year-1}_total',
            f'{year}_total',
            f'{year-1}_men',
            f'{year}_men',
            f'{year-1}_women',
            f'{year}_women']
    data.columns = col_list
    race_lis = [
                'White',
                'Black or African American',
                'Asian',
                'Hispanic or Latino ethnicity'
    ]


    df_cleaned = pd.DataFrame(columns = col_list)

    for i in range(0,4):
        if i <= 2:
            idx_first = data.index[data['occupation'] == race_lis[i]][0]
            idx_last = data.index[data['occupation'] == race_lis[i+1]][0]
            df_selected = data.loc[idx_first:idx_last,]
            df_selected['race'] = race_lis[i]
            df_cleaned = pd.concat([df_cleaned, df_selected])
        else:
            idx_first = data.index[data['occupation'] == race_lis[i]][0]
            df_selected = data.loc[idx_first: ,]
            df_selected['race'] = race_lis[i]
            df_cleaned = pd.concat([df_cleaned, df_selected])
    
    df_cleaned = df_cleaned.dropna()
    df_cleaned_year = df_cleaned[['occupation', f'{year}_total', f'{year}_men', f'{year}_women', 'race']]
    df_cleaned_year['year'] = year
    df_cleaned_year.columns =  ['occupation', 'total', 'men', 'women', 'race', 'year']
    df_final_03_to_10 = pd.concat([df_final_03_to_10, df_cleaned_year])
```

```python
df_final_03_to_10['total'] = df_final_03_to_10['total'].str.replace(',', '')
df_final_03_to_10['men'] = df_final_03_to_10['men'].str.replace(',', '')
df_final_03_to_10['women'] = df_final_03_to_10['women'].str.replace(',', '')
```

```python
df_final_03_to_10[['total', 'men', 'women', 'year']]  =  df_final_03_to_10[[ 'total', 'men', 'women', 'year']].apply(pd.to_numeric)
df_final_03_to_10.info()
```

```python
df_final_03_to_10.to_csv('data_2003_to_2010.csv')
```
